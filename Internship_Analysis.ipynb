```python
import pandas as pd
import random

names = [
    "Ali","Ahmed","Sara","Ayesha","Fatima","Usman","Hamza","Bilal",
    "Hina","Noor","Zain","Sana","Iqra","Areeba","Daniyal",
    "Saad","Laiba","Maham","Taha","Anas"
]

departments = ["IT","HR","Marketing","Finance","AI"]

durations = [2,3,4,5,6]

records = []

for i in range(100):

    student = random.choice(names) + " " + str(i+1)

    department = random.choice(departments)

    duration = random.choice(durations)

    mentor = random.randint(1,15)

    if mentor >= 8:
        status = "Completed"
    else:
        status = random.choice(["Completed","Dropped"])

    records.append([
        student,
        department,
        duration,
        mentor,
        status
    ])

df = pd.DataFrame(records, columns=[
    "Student",
    "Department",
    "Duration(Months)",
    "Mentor Meetings",
    "Status"
])

df.to_csv("internship_data.csv", index=False)

print("Dataset Created Successfully!")

df.head()
```

```python
data["Department"].value_counts().plot(
    kind="bar",
    figsize=(7,5)
)

plt.title("Students by Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")
plt.show()
```

```python
data["Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Internship Completion Rate")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    data["Mentor Meetings"],
    data["Duration(Months)"]
)

plt.title("Mentor Meetings vs Internship Duration")
plt.xlabel("Mentor Meetings")
plt.ylabel("Duration (Months)")

plt.show()
```
